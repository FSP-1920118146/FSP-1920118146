```python
# Implementasi Vigen√®re Cipher

def vigenere_encrypt(plain, key):
    plain = plain.upper().replace(" ", "")
    key = key.upper()
    result = ""
    for i in range(len(plain)):
        p = ord(plain[i]) - 65
        k = ord(key[i % len(key)]) - 65
        c = (p + k) % 26
        result += chr(c + 65)
    return result


def vigenere_decrypt(cipher, key):
    cipher = cipher.upper().replace(" ", "")
    key = key.upper()
    result = ""
    for i in range(len(cipher)):
        c = ord(cipher[i]) - 65
        k = ord(key[i % len(key)]) - 65
        p = (c - k) % 26
        result += chr(p + 65)
    return result


# Contoh uji
plaintext = "CRYPTOGRAPHY"
key = "LEMON"
ciphertext = vigenere_encrypt(plaintext, key)
decrypted = vigenere_decrypt(ciphertext, key)

print("Plaintext :", plaintext)
print("Key       :", key)
print("Ciphertext:", ciphertext)
print("Dekripsi  :", decrypted)
```

```python
from collections import Counter
import matplotlib.pyplot as plt

cipher = "NVKDGZKDOCSC"
count = Counter(cipher)

print("Analisis Frekuensi:")
for k, v in count.items():
    print(f"{k} : {v/len(cipher):.2f}")

plt.bar(count.keys(), [v/len(cipher) for v in count.values()])
plt.title("Distribusi Frekuensi Huruf pada Ciphertext")
plt.xlabel("Huruf")
plt.ylabel("Frekuensi Relatif")
plt.show()
```
